```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Switch off inline mode to export SHAP plot
%matplotlib notebook
```

```python
data=pd.read_excel('../Data/Model_Data_Rev1.xlsx')
```

```python
data.head(5)
```

```python
data.dropna(inplace=True, how='any')
```

```python
data.columns
```

```python
# use all features as X
X=data[['Active_Rig_Count', 'Offshore_Rig_Supply',
       'Offshore_Rig_Active',
       'Land_Rig_Active_Count', 'Crude_Price_2010_USD',
       'Adjusted_Nat_Gas_Price_2010', 'World_Oil_Cunsump',
       'World_Oil_Production']]
```

```python
y=data[['Offshore_Utilization_Rate']]
```

```python
y.shape
```

```python
bins = np.linspace(0, 398, 10)
y_binned = np.digitize(y, bins)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,random_state=42,stratify=y_binned) # can't stratify=y cuz sample size too small
```

```python
#from sklearn.preprocessing import StandardScaler
```

```python
#X_scaler=StandardScaler().fit(X_train)
#X_train_scaled=X_scaler.transform(X_train)
#X_test_scaled=X_scaler.transform(X_test)
```

```python
from xgboost import XGBRegressor
```

```python
xgb=XGBRegressor()
xgb.fit(X_train,y_train)
```

```python
xgb.score(X_train,y_train)
```

```python
xgb.score(X_test,y_test)
```

```python
# swith off inline mode to save SHAP plot
#%matplotlib
```

```python
import shap
```

```python
shap.initjs()
explainer = shap.TreeExplainer(xgb)
shap_values = explainer.shap_values(X_test)
```

```python
shap.summary_plot(shap_values, X.columns, plot_type="bar", show=False)
```
