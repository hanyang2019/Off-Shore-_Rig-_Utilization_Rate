```python
import pandas as pd
```

```python
trans_df = pd.read_excel('transposed_RC.xlsx', sheet_name='Sheet2')
```

```python
trans_df.head()
```

```python
land_df=trans_df[trans_df['Type']=='L']
land_df=land_df.drop('Type',axis=1)
land_df.set_index(['Unnamed: 0'], inplace=True)
trans_land=land_df.transpose()
headers=trans_land.columns.tolist()
df_list=[]
for column in headers:
    df=trans_land[[column]]
    df['nation']=column
    df.rename(columns={column:'Count'}, inplace=True)
    df_list.append(df)
new_land=pd.concat(df_list)
new_land.reset_index(drop=False, inplace=True)
new_land['Type']="In-Land"
new_land.rename(columns={'index':'Date','nation':'country'},inplace=True)
new_land
```

```python
shore_df=trans_df[trans_df['Type']=='O']
shore_df=shore_df.drop('Type',axis=1)
shore_df.set_index(['Unnamed: 0'], inplace=True)
trans_shore=shore_df.transpose()
headers=trans_shore.columns.tolist()
df_list=[]
for column in headers:
    df=trans_shore[[column]]
    df['nation']=column
    df.rename(columns={column:'Count'}, inplace=True)
    df_list.append(df)
new_shore=pd.concat(df_list)
new_shore.reset_index(drop=False, inplace=True)
new_shore['Type']="Off-Shore"
new_shore.rename(columns={'index':'Date','nation':'country'},inplace=True)
new_shore
```

```python
combine_df=pd.concat([new_land, new_shore])
```

```python
combine_df.to_csv('RC_data.csv',index=False,header=True)
```
